```python
import random

# Function to read words from a file
def read_words_from_file(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()

# Function to read characters from a file
def read_chars_from_file(file_path):
    with open(file_path, 'r') as file:
        return list(file.read().strip())

# Paths to the files containing words, digits, and symbols
words_file_path = 'top_words.txt'
digits_file_path = 'numbers.txt'
symbols_file_path = 'special_characters.txt'

# Read data from files
word_list = read_words_from_file(words_file_path)
DIGITS = read_chars_from_file(digits_file_path)
SYMBOLS = read_chars_from_file(symbols_file_path)

# Maximum length of password needed
MAX_LEN = 12

# Read data from files
word_list = read_words_from_file(words_file_path)
DIGITS = read_chars_from_file(digits_file_path)
SYMBOLS = read_chars_from_file(symbols_file_path)

# Update the symbols list to match the check_password_strength requirements
SYMBOLS = [symbol for symbol in SYMBOLS if symbol in "$@#"]

# Generate a password using a word from the word list
def generate_password():
    word = random.choice(word_list)
    # Convert some letters to uppercase for a mix of case
    word_mixed_case = ''.join(random.choice([k.upper(), k]) for k in word)
    # Shuffle the letters of the word
    shuffled_word = ''.join(random.sample(word_mixed_case, len(word_mixed_case)))
    # Insert a random digit and symbol at random positions in the shuffled word
    rand_index_digit = random.randint(0, len(shuffled_word))
    rand_index_symbol = random.randint(0, len(shuffled_word))
    # Ensure that the digit and symbol are not inserted at the same position
    while rand_index_digit == rand_index_symbol:
        rand_index_symbol = random.randint(0, len(shuffled_word))
    temp_password = (shuffled_word[:rand_index_digit] + random.choice(DIGITS) + 
                     shuffled_word[rand_index_digit:rand_index_symbol] + random.choice(SYMBOLS) + 
                     shuffled_word[rand_index_symbol:])
    # Adjust if the password exceeds the maximum length
    if len(temp_password) > MAX_LEN:
        temp_password = temp_password[:MAX_LEN]
    # Convert temporary password into array and shuffle
    temp_password_list = list(temp_password)
    random.shuffle(temp_password_list)
    # Join the shuffled characters to form the final password
    final_password = "".join(temp_password_list)
    return final_password


# Check password strength (function remains unchanged)
def check_password_strength(password):
    if len(password) < 8:
        return "Password is too short."
    if not any(char.isdigit() for char in password):
        return "Password should have at least one numeral."
    if not any(char.isupper() for char in password):
        return "Password should have at least one uppercase letter."
    if not any(char.islower() for char in password):
        return "Password should have at least one lowercase letter."
    if not any(char in SYMBOLS for char in password):
        return "Password should have at least one of the symbols $@#."
    return "Strong Password"

# Main loop to generate and check the password strength
while True:
    password = generate_password()
    if check_password_strength(password) == "Strong Password":
        break

print("Generated password: ", password)
```
