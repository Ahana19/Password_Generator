```python
import random

def read_words_from_file(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()

def read_chars_from_file(file_path):
    with open(file_path, 'r') as file:
        return list(file.read().strip())

words_file_path = 'dictionary.txt'
digits_file_path = 'numbers.txt'
symbols_file_path = 'special_characters.txt'
common_patterns_file_path = 'common_patterns.txt'  # Renamed for clarity

# Reading data from files
word_list = read_words_from_file(words_file_path)
DIGITS = read_chars_from_file(digits_file_path)
SYMBOLS = read_chars_from_file(symbols_file_path)

# Update the symbols list to match the check_password_strength requirements
SYMBOLS = [symbol for symbol in SYMBOLS if symbol in "$@#"]

# Read common password patterns from file
with open(common_patterns_file_path, 'r') as common_patterns_file:
    COMMON_PATTERNS = common_patterns_file.read().splitlines()

desired_length = int(input("Enter the desired password length (5-15): "))

# Function to generate a password with customizable length
def generate_password(length=desired_length):
    if length < 5 or length > 15:
        return "Password length must be more than 4 and less than 16."

    while True:
        word = random.choice(word_list)
        # Convert some letters to uppercase for a mix of case
        # word_mixed_case = ''.join(random.choice([k.upper(), k]) for k in word)
        # Insert a random digit and symbol at the end of the word
        temp_password = word + random.choice(SYMBOLS) + random.choice(DIGITS)

        # If the password does not meet the required length, add another word before the special character
        while len(temp_password) < length:
            # Find the remaining length needed
            remaining_length = length - len(temp_password) + 1  # +1 to replace the special character
            # Select another word that matches the remaining length
            additional_word = random.choice([word for word in word_list if len(word) == remaining_length])

            # Avoid common patterns
            if any(pattern in additional_word.lower() for pattern in COMMON_PATTERNS):
                continue

            temp_password += "-" + additional_word

        # Check if the password contains common patterns
        if any(pattern in temp_password.lower() for pattern in COMMON_PATTERNS) or len(temp_password) != length:
            continue
        else:
            return temp_password

# Example usage:
custom_password = generate_password()
print(f"Your custom password: {custom_password}")
```

```python
import random

def read_words_from_file(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()

def read_chars_from_file(file_path):
    with open(file_path, 'r') as file:
        return list(file.read().strip())

words_file_path = 'dictionary.txt'
digits_file_path = 'numbers.txt'
symbols_file_path = 'special_characters.txt'
common_patterns_file_path = 'common_patterns.txt'  # Renamed for clarity

# Reading data from files
word_list = read_words_from_file(words_file_path)
DIGITS = read_chars_from_file(digits_file_path)
SYMBOLS = read_chars_from_file(symbols_file_path)

# Update the symbols list to match the check_password_strength requirements
SYMBOLS = [symbol for symbol in SYMBOLS if symbol in "$@#"]

# Read common password patterns from file
with open(common_patterns_file_path, 'r') as common_patterns_file:
    COMMON_PATTERNS = common_patterns_file.read().splitlines()

desired_length = int(input("Enter the desired password length (5-15): "))

# Function to generate a passphrase-based password with customizable length
def generate_password(length=desired_length, personal_element=""):
    if length < 5 or length > 15:
        return "Password length must be more than 4 and less than 16."

    while True:
        # Generate a passphrase by concatenating multiple random words
        passphrase = '-'.join(random.choice(word_list) for _ in range(3))
        # Insert a random digit and symbol at the end of the passphrase
        temp_password = passphrase + random.choice(SYMBOLS) + random.choice(DIGITS)

        # Add the personal element if specified
        if personal_element:
            temp_password += "-" + personal_element

        # Check if the password contains common patterns
        if any(pattern in temp_password.lower() for pattern in COMMON_PATTERNS) or len(temp_password) != length:
            continue
        else:
            return temp_password

# Function to check password strength
def check_password_strength(password):
    # Define your criteria for password strength here
    # Example criteria:
    # 1. Minimum length (e.g., at least 8 characters)
    # 2. Mix of uppercase and lowercase letters
    # 3. Contains at least one digit
    # 4. Contains at least one special character
    # 5. Avoids common patterns (e.g., "password123", "qwerty")

    # Check minimum length
    if len(password) < 8:
        return False

    # Check uppercase and lowercase letters
    if not any(c.isupper() for c in password) or not any(c.islower() for c in password):
        return False

    # Check for at least one digit
    if not any(c.isdigit() for c in password):
        return False

    # Check for at least one special character
    if not any(c in "$@#" for c in password):
        return False

    # Avoid common patterns (customize this part based on your list of common patterns)
    common_patterns = ["password", "123456", "qwerty"]
    if any(pattern in password.lower() for pattern in common_patterns):
        return False

    # If all checks pass, the password is strong
    return True

# Example usage:
user_provided_password = input("Enter your own password: ")
if check_password_strength(user_provided_password):
    print("Your password meets the strength criteria.")
else:
    print("Your password does not meet the strength criteria. Please choose a stronger password.")


# Example usage:
custom_password = generate_password()
if check_password_strength(custom_password):
    print(f"Your custom password: {custom_password}")
else:
    print("Generated password does not meet the strength criteria. Please try again.")

# You can also prompt the user to input their own password and check its strength
user_provided_password = input("Enter your own password: ")
if check_password_strength(user_provided_password):
    print("Your password meets the strength criteria.")
else:
    print("Your password does not meet the strength criteria. Please choose a stronger password.")
```
